```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pymc as pm
import pandas as pd

sns.set()
```

```python
size = 100000

rate = 1/5
exp = pm.rexponential(beta=rate,size=size)
print (exp.mean(),exp.std())

_ = plt.hist(exp)
```

```python

#https://towardsdatascience.com/what-is-exponential-distribution-7bdd08590e2a

lambda_ = exp.mean()
poisson = pm.rpoisson(lambda_,size=size)
print (poisson.mean(),poisson.std())

_ = plt.hist(poisson)
```

```python
nr_samples = 200
sample_exp = np.random.choice(exp,replace=True,size=nr_samples)
sample_poisson_dist = pm.rpoisson(sample_exp,size=nr_samples)

_ = plt.hist(sample_poisson_dist)
```
