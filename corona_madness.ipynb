```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
pop = 8e9
infected = 1e5
dead = 4000
x_labels = ['world population','Corona Infected','Corona Dead']
plt.figure(figsize=(18,12))
plt.bar(np.arange(3),[pop,infected,dead])
plt.xticks(np.arange(3),x_labels)
plt.title('The Madness of Crowds')
#plt.yscale('log')
plt.ylabel('NOTE: Linear Scale!')
plt.savefig('corona_madness_lin.jpg',format='jpg')
```
