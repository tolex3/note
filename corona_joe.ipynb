```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

scrape_df = pd.read_pickle('scrape_df.pkl')

matrix_cols = 1000000

factor_1 = np.ones(matrix_cols)
factor_0 = factor_1 * 5.5
factor_2 = factor_1 * 1.5
factor_3 = factor_1 * 2
factor_4 = factor_1 * 2.5
factor_5 = factor_1 * 3
factor_6 = factor_1 * 3.5
factor_7 = factor_1 * 4
factor_8 = factor_1 * 4.5
factor_9 = factor_1 * 5

factor = np.concatenate([factor_1,factor_2,factor_3,factor_4,factor_5,
                        factor_0,factor_6,factor_7,factor_8,factor_9])

ratio_1 = np.linspace(0.001,0.1,matrix_cols)
ratio_2 = np.linspace(0.001,0.1,matrix_cols)
ratio_3 = np.linspace(0.001,0.1,matrix_cols)
ratio_4 = np.linspace(0.001,0.1,matrix_cols)
ratio_5 = np.linspace(0.001,0.1,matrix_cols)
ratio_6 = np.linspace(0.001,0.1,matrix_cols)
ratio_7 = np.linspace(0.001,0.1,matrix_cols)
ratio_8 = np.linspace(0.001,0.1,matrix_cols)
ratio_9 = np.linspace(0.001,0.1,matrix_cols)
ratio_0 = np.linspace(0.001,0.1,matrix_cols)


ratio = np.concatenate([ratio_1,ratio_2,ratio_3,ratio_4,ratio_5,
                       ratio_0,ratio_6,ratio_7,ratio_8,ratio_9])

arrays = np.array([factor,ratio])

def compute_params(country,confirmed,actual_dead):
    

    data = confirmed * arrays[1,:] * arrays[0,:]
    index = pd.MultiIndex.from_arrays(arrays,names=('factor','dead_pct'))


    df = pd.DataFrame(index=index,data=data,columns=['projected_dead'])
    df['actual_infected'] = arrays[0,:] * confirmed
    df = df[['actual_infected','projected_dead']]
    df['actual_dead'] = actual_dead
    df['confirmed'] = confirmed
    
    mask = ~(df['projected_dead'] < df['actual_dead'])
    df = df[mask]
    df['delta'] = df['projected_dead'] - df['actual_dead']
    min_delta = df['delta'].min()
    mask = df['delta'] == min_delta
    df = df[mask]
    df.reset_index(inplace=True)
    df.index = ([country] * len(df))
    return df


param_df = pd.DataFrame()

for country in scrape_df.index:
    confirmed = scrape_df.loc[country,'confirmed']
    dead = scrape_df.loc[country,'dead']
    temp_df = compute_params(country,confirmed,dead)
    param_df = pd.concat([param_df,temp_df],axis=0)
    
param_df.head(40)
```
