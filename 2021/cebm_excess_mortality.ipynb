```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set()
```

```python
url = 'https://www.cebm.net/covid-19/excess-mortality-across-countries-in-2020/?fbclid=IwAR3r3iivDSLQ2mdk6KXamarW2vpYyXM7LA2SN46EC3fNeJaoPTSoGDraQ1s'

tables = pd.read_html(url,header=0,index_col=0)
    
tables[0]
```

```python
datasource = 'DataSource : https://www.cebm.net/'
title = datasource
tables[0].plot(kind='bar',figsize=(18,12),subplots=True,title=title)
plt.savefig('cebm_excess_mortality.jpg',format='jpg')
```

```python
tables[0]['Percentage increase in Mortality per Age-Adjusted 100,000'] = \
tables[0]['Percentage increase in Mortality per Age-Adjusted 100,000'].apply(lambda x : x.strip('%')).astype(float)
```

```python
tables[0].to_excel('cebm_data.xlsx')
```

```python
title = '% Excess deaths calculated on Expectation avg. 2015-2019' + '\n' + datasource
tables[0]['Percentage increase in Mortality per Age-Adjusted 100,000'].sort_values(ascending=False).plot(
    kind='bar',figsize=(18,12),title=title)

plt.ylabel('percent excess deaths')
xtick_data = plt.xticks()
xtick_data[1][0].get_text()
xtick_data[1][0]

ax = plt.gca()
ax.xaxis.get_ticklabels()

plt.text(0,14.5, url,color='crimson')
plt.savefig('cebm_pct_excess_deaths.jpg',format='jpg')
```
