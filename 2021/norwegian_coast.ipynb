```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

import scipy.stats as sps

sns.set()
```

```python
#d = np.arange(1,7)
#n = np.array([25,13,9,7,5,5])

d = np.array([1,2,4,6,8,10])
n = np.array([100,25,9,4,4,1])

df = pd.DataFrame({'d' : d,'n' : n})

df.plot(x='d',y='n',style='o--')

```

```python
df['log_d'] = np.log10(df['d'])
df['log_n'] = np.log10(df['n'])

df
```

```python
df.plot(x='log_d',y='log_n',style='o--')
```

```python
slope,intercept,_,_,_ = sps.linregress(df['log_d'],df['log_n'])

print (slope)
print (intercept)

D = 1 - slope
print (D)
```

```python
ax = df.plot(x='log_d',y='log_n',style='o--')
ax.plot(df['log_d'],df['log_d'] * slope + intercept)
```
