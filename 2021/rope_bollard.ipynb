```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
### Rope aroung a ring or bollard - Effort Force calculation ###

### https://ocw.mit.edu/courses/physics/8-01sc-classical-mechanics-fall-2016/readings/MIT8_01F16_example8.11.pdf #

### https://youtu.be/hsbNusf2N8Y ###

# E = F * e ** (friction coeff (mu)  * angle around bollard (alpha))

# https://graphallthethings.com/posts/bollards/ #
mu = 0.4
F = 1

angle_deg = np.arange(0,3*360 + 90,90)
angle_rad = np.deg2rad(angle_deg)

E = F * np.exp(-mu * angle_rad)
print (1/E)

fig, axes = plt.subplots(2,sharex=True,figsize=(18,12))
axes[0].plot(angle_deg,E,'o--')
axes[0].grid(True)
axes[0].set_yticks(np.arange(0,1.1,0.1))
axes[0].set_ylabel('holding force fraction')


axes[1].set_xticks(np.arange(0,3*360+90,90))
axes[1].set_xticklabels(range(0,3*360+90,90),rotation=90)

axes[1].plot(angle_deg,1/E,'o--')
axes[1].grid(True)

axes[1].set_xlabel('contact angle [deg]')
axes[1].set_ylabel('holding force multiplier')

axes[0].set_yscale('log')
axes[1].set_yscale('log')

plt.suptitle(r'Effort Force for rope around a bollard [ friction coeff. $\mu$={} ]'.format(mu))
plt.savefig('rope_bollard.jpg',format='jpg')
```
